**변수별 결측치값 확인**

변수별 결측치값을 실제 결측치로 간주되도록 모두 `np.nan`으로 대체한다.
`missing_map는` 열별 결측치를 어레이로 지정한다.

```python
import re
delimiter = r' - |, ' # 결측값 표현에 사용된 구분자: ' - ' 또는 ', '

def parse_missing_codes(code_str):
    """'99, 999' -> [99.0, 999.0] 변환"""
    if pd.isna(code_str) or code_str == '':
        return np.array([np.nan])
    elif code_str == '없음':
        return np.array([np.nan])
    else:
        missing_range = [float(x.strip()) for x in re.split(delimiter, str(code_str))]
        return np.arange(missing_range[0], missing_range[-1] + 1)

# 변수명-결측값 어레이 사전 생성
missing_map = dict(zip(
    teacher_vars_df['변수'],
    teacher_vars_df['결측값'].apply(parse_missing_codes)
))
```

```python
missing_map
```

열별 결측값을 일괄적으로 `np.nan`으로 치환

```python
for col, codes in missing_map.items():
    if len(codes) > 1:
        teacher_df.loc[teacher_df.loc[:, col].isin(codes), col] = np.nan
```

### 훈련셋, 타깃 지정

```python
targets = ['FEEDBINSTR', 'ADAPTINSTR']
X = teacher_df.drop(columns=targets, axis=1)
y_feed = teacher_df['FEEDBINSTR']
y_adapt = teacher_df['ADAPTINSTR']
```

## RandomForest 학습

```python
# RandomForest 회귀 모델
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

### 하나의 RandomForest 학습

```python
# 1. Train/Test 분할
test_size = 0.2
random_state = 42
X_train, X_test, y_train_feed, y_test_feed = train_test_split(
    X, y_feed, test_size=test_size, random_state=random_state
)
print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# 2. 모델 정의 (초기 하이퍼파라미터)
rf_params = dict(
    n_estimators=300,
    max_depth=None,           # 깊이 제한 없음 (과적합시 줄이기)
    min_samples_split=2,
    min_samples_leaf=1,
    # max_features='sqrt',      # 일반적으로 좋은 기본값
    n_jobs=-1,
    random_state=random_state,
    oob_score=True,
)

rf_feed = RandomForestRegressor(**rf_params)

# 3. 학습
rf_feed.fit(X_train, y_train_feed)
# 4. 예측 및 기본 평가
y_pred_feed = rf_feed.predict(X_test)
rmse = mean_squared_error(y_test_feed, y_pred_feed)
mae = mean_absolute_error(y_test_feed, y_pred_feed)
r2 = r2_score(y_test_feed, y_pred_feed)
print(f"Test RMSE: {rmse:.4f}")
print(f"Test MAE : {mae:.4f}")
print(f"Test R2  : {r2:.4f}")

# 5. 교차검증 (R2, RMSE)
cv_r2 = cross_val_score(rf_feed, X_train, y_train_feed, cv=5, scoring='r2')
cv_rmse = -cross_val_score(rf_feed, X_train, y_train_feed, cv=5, scoring='neg_root_mean_squared_error')
print(f"CV R2 (mean±std): {cv_r2.mean():.4f} ± {cv_r2.std():.4f}")
print(f"CV RMSE (mean±std): {cv_rmse.mean():.4f} ± {cv_rmse.std():.4f}")
# 6. OOB Score (가능하면 일반화 성능 대략 확인)
if hasattr(rf_feed, 'oob_score_'):
    print(f"OOB R2: {rf_feed.oob_score_:.4f}")

# 7. 중요 변수 상위 20개 출력
importances = rf_feed.feature_importances_
indices = np.argsort(importances)[::-1][:20]
important_features = [(X.columns[i], importances[i]) for i in indices]
print("\nTop 20 Feature Importances:")
for name, score in important_features:
    print(f"{name:40s} {score:.4f}")
```

### 앙상블 학습: 모든 데이터 활용

```python
from sklearn.model_selection import RandomizedSearchCV

# 1. Train/Test 분할
test_size = 0.1
random_state = 42
X_train, X_test, y_train_feed, y_test_feed = train_test_split(
    X, y_feed, test_size=test_size, random_state=random_state
)

param_dist = {
    "n_estimators": [900, 1000, 1100],
    "max_depth": [40, 45, 50, 55],
    "min_samples_split": [2, 3, 4, 5],
    "min_samples_leaf": [2, 3, 4, 5],
    # "max_features": ["sqrt", "log2"],
}

rf_base = RandomForestRegressor(
    n_jobs=-1,
    random_state=42,
    oob_score=False,  # CV랑 중복되므로 끔
)

rnd_search = RandomizedSearchCV(
    rf_base,
    param_distributions=param_dist,
    n_iter=30,
    cv=3,
    scoring="r2",
    n_jobs=-1,
    random_state=42,
    verbose=1,
)

rnd_search.fit(X_train, y_train_feed)
print("Best R2 (CV):", rnd_search.best_score_)
print("Best params:", rnd_search.best_params_)

best_rf = rnd_search.best_estimator_
y_pred_best = best_rf.predict(X_test)
print("Test R2 (best):", r2_score(y_test_feed, y_pred_best))

importances = best_rf.feature_importances_
indices = np.argsort(importances)[::-1][:20]
important_features = [(X.columns[i], importances[i]) for i in indices]
print("\nTop 20 Feature Importances:")
for name, score in important_features:
    print(f"{name:40s} {score:.4f}")
```

### 앙상블 학습: 일부 특성만 활용

일부 특성만 대상으로할 경우엔 성능이 떨어진다.

```python
from sklearn.model_selection import RandomizedSearchCV

# 1. Train/Test 분할
test_size = 0.1
random_state = 42
X_train, X_test, y_train_feed, y_test_feed = train_test_split(
    X, y_feed, test_size=test_size, random_state=random_state
)

param_dist = {
    "n_estimators": [500, 600, 700],
    "max_depth": [40, 45, 50, 55, 60],
    "min_samples_split": [2, 3, 4, 5],
    "min_samples_leaf": [2, 3, 4],
    "max_features": ["sqrt"],
}

rf_base = RandomForestRegressor(
    n_jobs=-1,
    random_state=42,
    oob_score=False,  # CV랑 중복되므로 끔
)

rnd_search = RandomizedSearchCV(
    rf_base,
    param_distributions=param_dist,
    n_iter=30,
    cv=3,
    scoring="r2",
    n_jobs=-1,
    random_state=42,
    verbose=1,
)

rnd_search.fit(X_train, y_train_feed)
print("Best R2 (CV):", rnd_search.best_score_)
print("Best params:", rnd_search.best_params_)

best_rf_ = rnd_search.best_estimator_
y_pred_best = best_rf_.predict(X_test)
print("Test R2 (best):", r2_score(y_test_feed, y_pred_best))

importances = best_rf_.feature_importances_
indices_ = np.argsort(importances)[::-1][:20]
important_features_ = [(X.columns[i], importances[i]) for i in indices_]
print("\nTop 20 Feature Importances:")
for name, score in important_features_:
    print(f"{name:40s} {score:.4f}")
```
